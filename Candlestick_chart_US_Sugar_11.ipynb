```python
#Description: Thia program shows how to create an interactive candle stick chart in  Python using Plotly

#Import the libraries
import plotly.graph_objects as go
import pandas as pd
import numpy as np

```

```python
#Load the stock data
from google.colab import files
uploaded = files.upload()
```

```python
#Store the data onto a data frame
df = pd.read_csv('US Sugar #11 Futures Historical Data.csv')

#Set the index
df = df.set_index(pd.DatetimeIndex(df['Date'].values))

#Set up data in USD / MT
c = ["Price","Open", "High", "Low"]
m = 1120 / 50.8023454
df_mt = df[c] * m

#Set up data in XAF / MT
df_mt_xaf = df_mt[:] * 600

#Set up data in XAF / Sac
df_sac_xaf = df_mt_xaf[:] / 20
df_sac_xaf
```

```python
#1 Create an interactive candlestick chart USD / MT

figure_USD = go.Figure(
    data = [
            go.Candlestick(
                 x = df_mt.index,
                 low = df_mt['Low'],
                 high = df_mt['High'],
                 close = df_mt['Price'],
                 open = df_mt['Open'],
                 increasing_line_color = 'green',
                 decreasing_line_color = 'red'

            )
    ]
)

#figure.update_layout(xaxis_rangeslidervisible = False)

figure_USD.update_layout(
    title = 'US Sugar #11 Futures USD($)/MT [FOB] ',
    yaxis_title = 'Price  USD($) /  MT ',
    xaxis_title = 'Date'


)

figure_USD.show()

#2 Create an interactive candlestick chart XAF/ MT

figure_XAF = go.Figure(
    data = [
            go.Candlestick(
                 x = df_mt_xaf.index,
                 low = df_mt_xaf['Low'],
                 high = df_mt_xaf['High'],
                 close = df_mt_xaf['Price'],
                 open = df_mt_xaf['Open'],
                 increasing_line_color = 'green',
                 decreasing_line_color = 'red'

            )
    ]
)

#figure.update_layout(xaxis_rangeslidervisible = False)

figure_XAF.update_layout(
    title = 'US Sugar #11 Futures FCFA/MT [FOB]',
    yaxis_title = 'Price  FCFA /  MT ',
    xaxis_title = 'Date'


)

figure_XAF.show()

#3 Create an interactive candlestick chart XAF/ Sac

figure_XAF_Sac = go.Figure(
    data = [
            go.Candlestick(
                 x = df_sac_xaf.index,
                 low = df_sac_xaf['Low'],
                 high = df_sac_xaf['High'],
                 close = df_sac_xaf['Price'],
                 open = df_sac_xaf['Open'],
                 increasing_line_color = 'green',
                 decreasing_line_color = 'red'

            )
    ]
)

#figure.update_layout(xaxis_rangeslidervisible = False)

figure_XAF_Sac.update_layout(
    title = 'US Sugar #11 Futures FCFA/Sac [FOB]',
    yaxis_title = 'Price  FCFA /  Sac ',
    xaxis_title = 'Date'


)

figure_XAF_Sac.show()
```
